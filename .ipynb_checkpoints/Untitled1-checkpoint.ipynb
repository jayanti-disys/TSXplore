```python
from datetime import datetime, timedelta 
import pandas as  pd
import matplotlib.pyplot as plt
#from dateutil.relativedelta import relativedelta
#df = pd.read_csv(r"C:\Users\jayanti.prasad\Data\TS-DATA\airline\AirPassengers.csv",parse_dates=['Date'])

df = pd.read_csv(r"C:\Users\jayanti.prasad\Data\TS-DATA\covid-19-global.csv",parse_dates=['date'])
df = df.sort_values (by=['date'])


dfg  = df.groupby(['date']).sum()
dfg  = dfg.sort_values (by=['date'])




df1 = dfg.diff(1, axis=0)
dfg = df1 

dfg.to_csv("covid19.csv")


fig, axs = plt.subplots (3,1,figsize=(12,12))
axs[0].plot(dfg['confirmed'])
axs[1].plot(dfg['recovered'])
axs[2].plot(dfg['deaths'])

plt.show()


```
